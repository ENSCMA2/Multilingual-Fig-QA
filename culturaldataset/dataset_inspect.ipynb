## url search into c4


```python
from datasets import load_dataset

c4 = load_dataset("allenai/c4", LANG_CODE)
c4_dev_dict = c4.data['validation'].to_pydict()
c4_train_dict = c4.data['train'].to_pydict()
```

```python
for ex in hf_dataset['example'][:10]:
    for i in range(len(c4_train_dict['text'])):
        if ex == c4_train_dict['text'][i]:
            print(c4_train_dict['url'][i])
```
